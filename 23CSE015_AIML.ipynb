```python
# 1a) BFS
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = []
queue = []

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")
        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)


print("Following is the Breadth-First Search:")
bfs(visited, graph, '5')
```

```python
#1b) DFS
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

visited = set()

def dfs(visited, graph, node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)


dfs(visited, graph, 'A')
```

```python
# A* Algorithm
import sys
import numpy as np

def aStarAlgo(start_node, stop_node):
    open_set = set(start_node)
    closed_set = set()
    g = {}
    parents = {}

    g[start_node] = 0
    parents[start_node] = start_node

    while len(open_set) > 0:
        n = None

        for v in open_set:
            if n == None or g[v] + heuristic(v) < g[n] + heuristic(n):
                n = v

        if n == stop_node or Graph_nodes[n] == None:
            pass
        else:
            for (m, weight) in get_neighbors(n):
                if m not in open_set and m not in closed_set:
                    open_set.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n
                        if m in closed_set:
                            closed_set.remove(m)
                            open_set.add(m)
        if n == None:
            print('Path does not exist!')
            return None

        if n == stop_node:
            path = []
            while parents[n] != n:
                path.append(n)
                n = parents[n]
            path.append(start_node)
            path.reverse()
            print('Path found: {}'.format(path))
            return path

        open_set.remove(n)
        closed_set.add(n)

    print('Path does not exist!')
    return None

def get_neighbors(v):
    if v in Graph_nodes:
        return Graph_nodes[v]
    else:
        return None

def heuristic(n):
    H_dist = {
        'A': 11,
        'B': 6,
        'C': 5,
        'D': 7,
        'E': 3,
        'F': 6,
        'G': 5,
        'H': 3,
        'I': 1,
        'J': 0
    }
    return H_dist[n]


Graph_nodes = {
    'A': [('B', 6), ('F', 3)],
    'B': [('A', 6), ('C', 3), ('D', 2)],
    'C': [('B', 3), ('D', 1), ('E', 5)],
    'D': [('B', 2), ('C', 1), ('E', 8)],
    'E': [('C', 5), ('D', 8), ('I', 5), ('J', 5)],
    'F': [('A', 3), ('G', 1), ('H', 7)],
    'G': [('F', 1), ('I', 3)],
    'H': [('F', 7), ('I', 2)],
    'I': [('E', 5), ('G', 3), ('H', 2), ('J', 3)],
}

aStarAlgo('A', 'J')
```

```python
# 4th Bayesian Networks
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination


med_model = DiscreteBayesianNetwork([
    ('Smoking', 'LungDisease'),
    ('AirPollution', 'LungDisease'),
    ('LungDisease', 'Cough'),
    ('LungDisease', 'XRayResult')
])


cpd_smoking = TabularCPD(
    variable='Smoking',
    variable_card=2,
    values=[[0.7], [0.3]]
)

cpd_pollution = TabularCPD(
    variable='AirPollution',
    variable_card=2,
    values=[[0.6], [0.4]]
)

cpd_disease = TabularCPD(
    variable='LungDisease',
    variable_card=2,
    values=[[0.9, 0.7, 0.6, 0.1],
            [0.1, 0.3, 0.4, 0.9]],
    evidence=['Smoking', 'AirPollution'],
    evidence_card=[2, 2]
)

cpd_cough = TabularCPD(
    variable='Cough',
    variable_card=2,
    values=[[0.8, 0.1],
            [0.2, 0.9]],
    evidence=['LungDisease'],
    evidence_card=[2]
)

cpd_xray = TabularCPD(
    variable='XRayResult',
    variable_card=2,
    values=[[0.9, 0.2],
            [0.1, 0.8]],
    evidence=['LungDisease'],
    evidence_card=[2]
)


med_model.add_cpds(cpd_smoking, cpd_pollution, cpd_disease, cpd_cough, cpd_xray)


print("Model verification:", med_model.check_model())


med_infer = VariableElimination(med_model)


print("\nProbability distribution, P(Smoking):")
print(cpd_smoking)
print("\nProbability distribution, P(AirPollution):")
print(cpd_pollution)
print("\nJoint probability distribution, P(LungDisease | Smoking, AirPollution):")
print(cpd_disease)
print("\nJoint probability distribution, P(Cough | LungDisease):")
print(cpd_cough)
print("\nJoint probability distribution, P(XRayResult | LungDisease):")
print(cpd_xray)


print("\n1. Probability of LungDisease given Cough and XRay abnormal:")
q1 = med_infer.query(['LungDisease'], evidence={'Cough': 1, 'XRayResult': 1})
print(q1)

print("\n2. Probability of Smoking given XRay abnormal:")
q2 = med_infer.query(['Smoking'], evidence={'XRayResult': 1})
print(q2)
```

```python
# 5th Regression Model
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


data = pd.read_csv("/content/StressLevelDataset.csv")


data.fillna(data.mean(), inplace=True)


X = data.drop('stress_level', axis=1)  # Features
y = data['stress_level']               # Target variable


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


reg_model = LinearRegression()
reg_model.fit(X_train, y_train)


y_pred = reg_model.predict(X_test)


print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
print("R-squared (R²) Score:", r2_score(y_test, y_pred))


plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r', linewidth=2)
plt.title('Actual vs Predicted Stress Levels')
plt.xlabel('Actual Stress Level')
plt.ylabel('Predicted Stress Level')
plt.grid(True)
plt.show()


coefficients = pd.DataFrame(reg_model.coef_, X.columns, columns=['Coefficient'])
print("\nModel Coefficients:\n", coefficients)
```

```python
# 7th SVM
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from mlxtend.plotting import plot_decision_regions


data = pd.read_csv('/content/StressLevelDataset.csv')


feature1 = 'anxiety_level'
feature2 = 'sleep_quality'
feature3 = 'self_esteem'
target = 'stress_level'


X = data[[feature1, feature2, feature3]]
y = data[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


svm_model = svm.SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train_scaled, y_train)


y_pred = svm_model.predict(X_test_scaled)


print("\nModel Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))


plt.figure(figsize=(12, 6))

# Plot the training data
plt.subplot(1, 2, 1)
plt.scatter(X_train_scaled[y_train == 0, 0], X_train_scaled[y_train == 0, 1],
            color='blue', label='Stress Level 0', alpha=0.5)
plt.scatter(X_train_scaled[y_train == 1, 0], X_train_scaled[y_train == 1, 1],
            color='green', label='Stress Level 1', alpha=0.5)
plt.scatter(X_train_scaled[y_train == 2, 0], X_train_scaled[y_train == 2, 1],
            color='red', label='Stress Level 2', alpha=0.5)


ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()


xx = np.linspace(X_train_scaled[:, 0].min(), X_train_scaled[:, 0].max(), 30)
yy = np.linspace(X_train_scaled[:, 1].min(), X_train_scaled[:, 1].max(), 30)
zz = np.linspace(X_train_scaled[:, 2].min(), X_train_scaled[:, 2].max(), 30)  # Add z-dimension


YY, XX, ZZ = np.meshgrid(yy, xx, zz)  # Create 3D meshgrid


xy = np.vstack([XX.ravel(), YY.ravel(), ZZ.ravel()]).T  # Include all 3 features


Z = svm_model.decision_function(xy)



if len(svm_model.classes_) == 2:
    Z = Z.reshape(XX.shape)
else:

    Z = Z.reshape(XX.shape[0], XX.shape[1], XX.shape[2], -1)
    Z = np.max(Z, axis=3)


ax.contour(XX[:, :, 0], YY[:, :, 0], Z[:, :, 0], colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])


plt.xlabel(f'{feature1} (scaled)')
plt.ylabel(f'{feature2} (scaled)')
plt.title('SVM Decision Boundary (Training Data)')
plt.legend()


plt.subplot(1, 2, 2)
plt.scatter(X_test_scaled[y_pred == 0, 0], X_test_scaled[y_pred == 0, 1],
            color='blue', label='Predicted 0', alpha=0.5)
plt.scatter(X_test_scaled[y_pred == 1, 0], X_test_scaled[y_pred == 1, 1],
            color='green', label='Predicted 1', alpha=0.5)
plt.scatter(X_test_scaled[y_pred == 2, 0], X_test_scaled[y_pred == 2, 1],
            color='red', label='Predicted 2', alpha=0.5)


misclassified = (y_pred != y_test)
plt.scatter(X_test_scaled[misclassified, 0], X_test_scaled[misclassified, 1],
            color='black', marker='x', s=100, label='Misclassified')


ax = plt.gca()
ax.contour(XX[:, :, 0], YY[:, :, 0], Z[:, :, 0], colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

plt.xlabel(f'{feature1} (scaled)')
plt.ylabel(f'{feature2} (scaled)')
plt.title('SVM Predictions (Test Data)')
plt.legend()

plt.tight_layout()
plt.show()


plt.subplot(1, 2, 2)
ax3d = plt.axes(projection='3d')


scatter = ax3d.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], X_train_scaled[:, 2],
                       c=y_train, cmap='viridis', alpha=0.5)


x_range = np.linspace(X_train_scaled[:, 0].min(), X_train_scaled[:, 0].max(), 30)
y_range = np.linspace(X_train_scaled[:, 1].min(), X_train_scaled[:, 1].max(), 30)
X1, X2 = np.meshgrid(x_range, y_range)
X3 = (-svm_model.intercept_[0] - svm_model.coef_[0][0] * X1 - svm_model.coef_[0][1] * X2) / svm_model.coef_[0][2]


ax3d.plot_surface(X1, X2, X3, alpha=0.5, rstride=100, cstride=100, color='orange')

ax3d.set_xlabel(f'{feature1} (scaled)')
ax3d.set_ylabel(f'{feature2} (scaled)')
ax3d.set_zlabel(f'{feature3} (scaled)')
ax3d.set_title('SVM Decision Surface (3D)')

plt.tight_layout()
plt.show()
```

```python
#9th Clustering algorithms

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D


df = pd.read_csv('/content/StressLevelDataset.csv')
print("Dataset Loaded\n")
print("First 5 columns:")
print(df.iloc[:, :5].head())


df.dropna(inplace=True)


label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)


plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='red')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.grid()
plt.show()


k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df.iloc[:, 0], y=df.iloc[:, 1], hue=df['Cluster'], palette='Set1')
plt.title('K-Means Clustering (2D View)')
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.show()


print("\nCluster distribution:")
print(df['Cluster'].value_counts())

print("\nCluster-wise means (first 5 features):")
print(df.groupby('Cluster').mean().iloc[:, :5])



plt.figure(figsize=(7, 5))
sns.countplot(x='Cluster', data=df, palette='pastel')
plt.title('Number of Samples in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.grid(True)
plt.show()



cluster_means = df.groupby('Cluster').mean()
cluster_means.iloc[:, :3].plot(kind='bar', figsize=(10, 6))
plt.title("Cluster Centers (Mean of First 3 Features)")
plt.xlabel("Cluster")
plt.ylabel("Mean Value")
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
#10th EM for Bayesian Networks
import pandas as pd
import numpy as np
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import BayesianEstimator
from pgmpy.inference import VariableElimination


data = pd.read_csv('/content/StressLevelDataset.csv')


if 'sn' in data.columns:
    data = data.drop(columns=['sn'])


data = data.replace('?', np.nan)


for col in data.columns:
    data[col] = pd.to_numeric(data[col], errors='ignore')
    if pd.api.types.is_numeric_dtype(data[col]):

        data[col] = pd.Categorical(data[col])


model = DiscreteBayesianNetwork([
    ('anxiety_level', 'stress_level'),
    ('sleep_quality', 'stress_level'),
    ('study_load', 'stress_level'),
    ('peer_pressure', 'stress_level'),
    ('bullying', 'stress_level')
])


model.fit(data, estimator=BayesianEstimator)


inference = VariableElimination(model)


print("Possible values for anxiety_level:", data['anxiety_level'].cat.categories.tolist())
print("Possible values for sleep_quality:", data['sleep_quality'].cat.categories.tolist())
print("Possible values for stress_level:", data['stress_level'].cat.categories.tolist())


print("\nExample 1: Moderate anxiety (5) and average sleep quality (3)")
result = inference.query(
    variables=['stress_level'],
    evidence={'anxiety_level': 5, 'sleep_quality': 3}
)
print(result)


print("\nExample 2: High anxiety (10) and poor sleep quality (1)")
result = inference.query(
    variables=['stress_level'],
    evidence={'anxiety_level': 10, 'sleep_quality': 1}
)
print(result)


print("\nExample 3: Low anxiety (2) and good sleep quality (5)")
result = inference.query(
    variables=['stress_level'],
    evidence={'anxiety_level': 2, 'sleep_quality': 5}
)
print(result)


def query_stress_level(anxiety_val, sleep_val):
    """Helper function to query stress level given evidence"""
    try:
        result = inference.query(
            variables=['stress_level'],
            evidence={'anxiety_level': anxiety_val, 'sleep_quality': sleep_val}
        )
        print(f"\nStress level distribution given anxiety={anxiety_val}, sleep={sleep_val}:")
        print(result)
        return result
    except Exception as e:
        print(f"Error in query: {e}")
        print("Please use valid values from the categories shown above")


query_stress_level(8, 2)
```

```python
#12th Deep Learning NN Model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


df = pd.read_csv('/content/StressLevelDataset.csv')


print("Original Columns:", df.columns)


df.columns = df.columns.str.strip().str.lower()


df.replace('?', np.nan, inplace=True)
df = df.dropna()


le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col].astype(str))


X = df.drop('stress_level', axis=1)
y = df['stress_level']


if y.nunique() <= 10:
    y = le.fit_transform(y)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))


if len(np.unique(y)) > 2:
    model.add(Dense(len(np.unique(y)), activation='softmax'))
    model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
else:
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])


history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.1)


loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.2f}")


plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)


plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
